 ## 미국 대통령 신장(키) 데이타를 분석 
 
 
        ` 전체적인 기본 통계량
        ` 평균신장
        ` 그래프로 시각화

```python
import pandas as pd

data = pd.read_csv('data/president_heights.csv')
data
```

```python
# 기본통계량 (평균 / 표준편차 / 최솟값 / 최댓값 / 중앙값)
# 키에 대한 기본통계량
da = data['height(cm)']

print(da.mean())
print("-" * 50)
print(da.std())
print("-" * 50)
print(da.min())
print("-" * 50)
print(da.max())
print("-" * 50)
print(da.median())
```

```python
# 기본 통계량
data.info()    # 자료형만 보는

data.describe()
```
