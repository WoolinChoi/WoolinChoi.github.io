##  파산한 은행 데이타집합에서 분기별 은행수를 계산하기
[도서] Do it 데이타분석을 위한 판다스 입문

```python
import pandas as pd
from datetime import datetime as dt
import numpy as np

banks = pd.read_csv('./data/banklist.csv')
banks.head()
```

### (1) 데이터 구조 파악

```python
banks.info()
```

```python
banks.dtypes
```

### (2) Closing Date를 datetime 오브젝트로 변경

```python
banks['cd_parse'] = pd.to_datetime(banks['Closing Date'])
banks.head()
```

### (3) 날짜 정보에서 dt 접근자를 이용하여 연도(year)와 분기(quarter) 구하기

```python
d1 = banks['cd_parse'].dt
banks['year'] = d1.year

banks['quarter'] = np.where(d1.month<=3, 1, 
                                    np.where(d1.month<=6, 2,
                                         np.where(d1.month<=9, 3, 4)))
banks.tail()
```

### (4) 연도별 파산은행 수 구하기

```python
bank_year = banks.groupby('year').count()['Bank Name']
bank_year
```

### (5) 연도별, 분기별 파산은행 수 구하기

```python
bank_yq = banks.groupby(['year', 'quarter']).count()
bank_yq
```

### 결과 시각화 (간단하게)

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

fig = plt.figure(figsize=(100,30))
axes = plt.subplot()
axes = sns.barplot(x = bank_yq.index, y = bank_yq['Bank Name'])
axes.set_xticklabels(labels=bank_yq.index,rotation=40, fontsize=40)
axes.set_title('bankruptcy by year&quarter', fontsize=100)
```
