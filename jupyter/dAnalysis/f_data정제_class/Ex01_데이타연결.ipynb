## 1. 데이타 가져오기

```python
import pandas as pd

person = pd.read_csv('./data/survey_person.csv')
person.columns = ['person', 'personal', 'family']
person
```

```python
survey = pd.read_csv('./data/survey_survey.csv')
survey.columns = ['taken', 'person', 'quant', 'reading']
survey
```

```python
visited = pd.read_csv('./data/survey_visited.csv')
visited.columns = ['taken', 'site', 'dated']
visited
```

```python
site = pd.read_csv('./data/survey_site.csv')
site.columns = ['site', 'lat', 'long']
site
```

## 2. 데이타 연결하기
### concat 메소드 연결

```python
# redata = pd.concat([site, visited])  # 여러 데이터를 연결할때는 concat
# redata  # NaN값들이 발생
```

### Pandas의 merge()

```python
pd.merge(site, visited, left_on='site', right_on='site')
```

### DataFrame의 merge()

```python
site.merge(visited, left_on='site', right_on='site')
```

```python
site.merge(visited, left_on='site', right_on='site', how='outer')
```

### person과 survey 데이터셋 합치기

```python
pd.merge(person, survey, left_on='person', right_on='person')
```

### person과 visited 데이터셋 합치기

```python
per = person.merge(survey, left_on='person', right_on='person')
per.merge(visited, left_on='taken', right_on='taken')
```
