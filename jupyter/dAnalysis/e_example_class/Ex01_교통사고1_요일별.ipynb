### 교통사고 데이타셋에서 요일별 사망자수 필터링 + 시각화

(1) 요일별 사망자수 필터링과 합계 (groupby() 이용, sum() 추가)

(2) 그래프

```python
%matplotlib notebook

import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 한글깨짐처리
font_loc = "c:/Windows/Fonts/malgun.ttf"
font_name=font_manager.FontProperties(fname=font_loc).get_name()
rc('font', family=font_name)
```

```python
# 데이타 읽어오기
data=pd.read_csv('data/accidentdata.csv')
data
```

```python
# 요일별 사망자수의 합계 (sum() 추가)
yoil = data['사망자수'].groupby(data['요일']).sum()
yoil_data = yoil.reindex(index=['월','화','수','목','금','토','일'])
yoil_data
```

```python
# 그래프
plt.bar(yoil_data.index, yoil_data)
plt.title('요일별 사망자수 그래프')
plt.xlabel('요일')
plt.ylabel('사망자 수')
```
