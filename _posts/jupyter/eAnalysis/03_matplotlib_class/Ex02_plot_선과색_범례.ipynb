## 0. 플롯 스타일

```python
%matplotlib inline
    
import matplotlib.pyplot as plt
import numpy as np

# plt.style.use('classic')
plt.style.use('seaborn-whitegrid')

x = np.linspace(0, 10, 100)    # 0~10 사이의 수를 균등하게 100 추출
plt.plot(np.sin(x))
```

## 1. 색상지정

```python
plt.plot(np.sin(x-0), color='blue')               # 색상이름
plt.plot(np.sin(x-1), color='g')                  # 색상코드 
plt.plot(np.sin(x-2), color='#FFDD33')            # 16진수
plt.plot(np.sin(x-3), color='chartreuse')         # html 색상이름
plt.plot(np.sin(x-4), color='0.7')                # 0 ~ 1 회색
plt.plot(np.sin(x-5), color=(1.0, 0.2, 0.5))      # RGB 튜플형식
```

## 2. 선스타일

```python
# plt.plot(np.sin(x-0), linestyle='solid')
# plt.plot(np.sin(x-1), linestyle='dashed')
# plt.plot(np.sin(x-2), linestyle='dotted')
# plt.plot(np.sin(x-3), linestyle='dashdot')

plt.plot(np.sin(x-0), linestyle='-')
plt.plot(np.sin(x-1), linestyle='--')
plt.plot(np.sin(x-2), linestyle=':')
plt.plot(np.sin(x-3), linestyle='-.')
```

## 3. 선과 스타일을 간결하게 (대략 읽는 수준으로 )
### r(red) g(green) b(blue) c(cyan) m(magenta) y(yellow) k(black) 

```python
# solid green
plt.plot(np.sin(x-0), 'g-')

# dashed cyan
plt.plot(np.sin(x-1), 'c--')

# dashdot black
plt.plot(np.sin(x-2), 'k-.')

# dotted red
plt.plot(np.sin(x-3), 'r:')
```

## 4. 범례

```python
plt.plot(np.sin(x-0), 'g-', label='first')
plt.plot(np.sin(x-1), 'c--', label='second')
plt.plot(np.sin(x-2), 'k-.', label='thrid')
plt.plot(np.sin(x-3), 'r:', label='fourth')
plt.legend(loc=4)
```
