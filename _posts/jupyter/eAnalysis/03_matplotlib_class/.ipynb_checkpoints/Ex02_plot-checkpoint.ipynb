```python
# 기본적인 그래프
%matplotlib inline

import matplotlib.pyplot as plt

data = [10, 15, 19, 30, 20]     # y축
plt.plot(data)     # x축은 인덱스로 자동
```

- 그래프 틀(영역)을 만들기
   > plt.figure()
- 그래프 그리기
   > plt.plot(data)     
- 그래프 보여주기
   > plt.show()         

### *그래프 출력시 객체표시 같은 거 안 출력하게 하려면 뒤에 세미콜론(;) 붙이면 된다*

```python
plt.figure(figsize=(12, 8))    # figsize: 그래프 크기
plt.plot(data)
plt.show()
```

```python
# x축과 y축의 라벨 추가
plt.figure(figsize=(12, 8))
plt.plot(data)
plt.title('My Sample', fontsize=22)    # fontsize: 글씨 크기
plt.xlabel('data')
plt.ylabel('value')
plt.grid()    # gird: 격자표시
plt.show()
```

```python
# y축 데이타에 x축 데이타를 지정
step = ['one', 'two', 'three', 'four', 'five']
plt.plot(step, data);    # step: x축, data: y축
```

```python
# 속성 : linestyle , color , marker
plt.plot(step, data, linestyle='dashed', color='red', marker='o')
```

```python
# 속성 축약
plt.plot(step, data, 'go--')    # 속성축약 '색, 표시, 선'
```

[참고] 선의 두께와 선의 색을 지정
- lw의 기본값은 1
- 선 색
    { 'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w' }

```python
# 마커속성 
plt.plot(step, data, linestyle='dashed', color='red', 
         marker='o', markerfacecolor='cyan', markersize=20)
```
