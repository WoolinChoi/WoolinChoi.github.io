## groupby()


#### groupby()와 함께 사용하는 집계 메소드

- count : 누락값을 제외한 데이타 수를 반환
- size : 누락값을 포함한 데이타 수를 반환
- sum
- mean
- std
- var
- min
- max
- quantile(q=0.25)
- quantile(q=0.50)
- quantile(q=0.75)
- describe
- first
- last
- nth   : n번째 행 반환
- sem   : 평균의 표준편차 반환

```python
import pandas as pd

df = pd.read_excel('data/고속버스.xlsx') 
df
```

```python
# 그룹화 - '선별' 단위로 각각의 노선에 대한 총 갯수
dd = df.groupby('선별')
dd['선별'].count()
```

```python
# 정렬해서 출력
dd['선별'].count().sort_values(ascending=False)    # 내림차순
```

```python
%matplotlib inline

# 한글처리
from matplotlib import rc
rc('font', family='Malgun Gothic')
```

```python
# 노선별 수에 대한 그래프

dd['선별'].count().plot(kind='bar')
```

```python
# 차종별, 선별별 그룹화 작업 + 그래프
premium = df[df['차종'] == '우등'].groupby('선별').count()['총운행횟수']
express = df[df['차종'] == '고속'].groupby('선별').count()['총운행횟수']

result = pd.DataFrame({'선별':premium.index, '우등' :premium.values, '고속':express.values})
result.plot(x='선별', y=['우등', '고속'], kind="bar")
```

```python
# 총이용인원에 대한 평균
tmp3 = df.groupby(df['선별'])[['총이용인원', '총운행횟수', '이용율', '거리']].mean()
tmp3.plot(kind='bar', figsize=(12, 5), title='노선별 정보')

agg_format = {'총이용인원':'sum', '이용율':'mean'}

tmp4 = df.groupby(df['선별'])[['총이용인원', '이용율']].agg(agg_format)
tmp4.plot(kind='bar',figsize=(12, 5), title='노선별 정보', y=['총이용인원', '이용율'], subplots=True)
```

```python
# 교차테이블(cross tab) - 엑셀의 Pivot 유사
result = pd.crosstab(df['차종'], df['선별'])
result
```

```python
result2 = pd.crosstab(df['차종'], df['선별'], margins=True)
result2
```

```python
result.plot(kind='bar')
```

```python
# '고속'에서 선별을 그래프 (pie)
result.loc['고속'].plot(kind='pie')
```

```python
# 고속과 우등을 합친 전체에서 노선별 그래프 (pie)
result3 = result.loc['우등'] + result.loc['고속']
result3.plot(kind='pie')
```
